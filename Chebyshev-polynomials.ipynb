# Chebyshev polynomials

Author: Wouter Kouw. Last update: 21-feb-2022

The Chebyshev polynomials $T_n(x)$ are polynomials with the largest possible leading coefficient, whose absolute value on the interval [−1, 1] is bounded by 1. 

<img align="left" width=800pt src="https://upload.wikimedia.org/wikipedia/commons/thumb/5/5c/Chebyshev_Polynomials_of_the_First_Kind.svg/561px-Chebyshev_Polynomials_of_the_First_Kind.svg.png">

They form an orthonormal basis. A function defined on the domain $-1 \leq x \leq 1$ can be expressed by the series:

$$ f(x) = \sum_{n=0}^{\infty} \alpha_n T_n(x) \, .$$

A finite approximation is a Chebyshev series.

```julia
using Pkg
Pkg.activate(".")
Pkg.instantiate()
IJulia.clear_output();
```

```julia
using LinearAlgebra
using LaTeXStrings
using Distributions
using Polynomials
using Plots
default(grid=false, label="", size=(800,400), margin=10Plots.pt, labelfontsize=12, legendfontsize=12)
```

```julia
# Trial Polynomial
p = Polynomial([1, -1, 1, 1])

# Trial Chebyshev
q = ChebyshevT([1, -1, 1, 1])

# Plot
xx = range(-1, stop=1, step=0.01)
plot(xx, p.(xx), color="red", lw=3, label="standard polynomial", xlabel="x", ylabel="f(x)")
plot!(xx, q.(xx), color="blue", lw=3, label="Chebyshev pol.")
```

```julia
# Max order
K = 5

# Plot
plot(xlabel="x", ylabel="f(x)")

for k = 1:K
    
    # Trial Chebyshev
    q = ChebyshevT(ones(k))
    
    # Plot
    plot!(xx, q.(xx), lw=3, label="k = "*string(k), legend=:topleft)
end
plot!()
```

```julia
savefig("figures/Chebyshev-pols-uptoK"*string(K)*".png")
```

```julia
# Generate random coefficients
N = 10
K = 4
Θ = rand(N,K)*2 .-1

# Plot
plot(xlabel="x", ylabel="f(x)")

for n = 1:10
    
    # Trial Chebyshev
    q = ChebyshevT(Θ[n,:])
    
    # Plot
    plot!(xx, q.(xx), lw=3)
end
plot!()
```
